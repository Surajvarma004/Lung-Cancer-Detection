# Dependent and independent Variable

```python
X = data.iloc[:,:-1]
```

```python
X
```

```python
y = data['LUNG_CANCER']
```

```python
y
```

# Spliting the Data into Training data and Testing data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=3,test_size=0.2)
```

# Scaling the data

```python
 from sklearn.preprocessing import MinMaxScaler
```

```python
scale=MinMaxScaler()
# Scaling the training data 
X_train_scaled=pd.DataFrame(scale.fit_transform(X_train),columns=X_train.columns)
X_train_scaled
```

```python
# Scaling the testing data
X_test_scaled = pd.DataFrame(scale.fit_transform(X_test),columns=X_test.columns)
X_test_scaled
```

# Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(X_train,y_train)
```

# Prediction

```python
prediction = model.predict(X_test) # Predicting the testing data
```

```python
prediction
```

# Checking the accuracy 

```python
from sklearn import metrics
```

```python
accuracy = metrics.accuracy_score(y_test,prediction)
accuracy
```

The model is 91.93% accurate

# Ploting confusion metrix 

```python
from sklearn.metrics import confusion_matrix
```

```python
pred_list = [model]

for i in pred_list:
    print("Score : ",i.score(X_test,y_test))
    y_pred = i.predict(X_test)
    sns.heatmap(confusion_matrix(y_test,y_pred),annot = True)
    plt.xlabel("Y_pred")
    plt.ylabel("Y_test")
    plt.title(i)
    plt.show()
```

This is plot of confusion metrix of y_test and y_prediction using RandomForestClassifier
